# 🧠 Introduction to Python Lists

Lists are one of the **most important data structures in Python**.

A **list** allows us to store **multiple values in a single variable**.

Unlike basic variables that store one value, lists can store **collections of data**.

For example:

- a list of numbers
- a list of student names
- a dataset of images
- a collection of machine learning features

Lists are heavily used in **data science and AI** because datasets often contain **thousands or millions of values**.

### Lists are useful for:

- 📊 storing datasets  
- 🤖 machine learning inputs  
- 📂 managing collections of data  
- 🔁 iterating through items using loops  

Today we will learn:

- 📦 Creating lists  
- 🔎 List indexing  
- ✂️ List slicing  
- ⚙️ List methods

# 📦 Creating a List

A list is created using **square brackets `[ ]`**.

Multiple elements are separated by **commas**.

### Example

```python
numbers = [10, 20, 30, 40, 50]
print(numbers)
```

Output:
```
[10, 20, 30, 40, 50]
```
Lists can store different data types:
```python
data = [10, "AI", 3.14, True]
print(data)
```
Output:
```
[10, 'AI', 3.14, True]
```

````python
numbers = [10, 20, 30, 40, 50]
print(numbers)
````

#🔎 Accessing List Elements (Indexing)

Each element in a list has a position called an index.

Important rule:

Python list indexing starts from 0

Example list:

Index:   0   1   2   3

Values: [10, 20, 30, 40]

Example
```python
numbers = [10, 20, 30, 40]

print(numbers[0])
print(numbers[2])
```
Output:
```
10
30
```
###Negative Indexing

Python also allows negative indexing.

-1 → last element

-2 → second last element

Example:
```python
numbers = [10, 20, 30, 40]
print(numbers[-1])
```
Output:
```
40
```

````python
numbers = [10, 20, 30, 40]

print(numbers[0])
print(numbers[2])
````

#✂️ List Slicing

Slicing allows us to extract a portion of a list.

Syntax
```
list[start : end]
```

###Important rule:

The end index is not included

Example
```python
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])
```
Output:
```
[20, 30, 40]
```

Explanation:
* start → index 1
* end → index 4 (not included)

###Other Slicing Examples
```
numbers = [10, 20, 30, 40, 50]

print(numbers[:3])
print(numbers[2:])
```

Output:
```
[10, 20, 30]
[30, 40, 50]
```

````python
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])
````

#⚙️ Useful List Methods

Python provides many built-in methods to modify lists.

##➕ Adding Elements
###append()

Adds an element to the end of the list.
```python
numbers = [1, 2, 3]
numbers.append(4)
print(numbers)
```

Output:
```
[1, 2, 3, 4]
```

###insert()

Adds an element at a specific position.
```python
numbers = [1, 2, 3]
numbers.insert(1, 10)
print(numbers)
```

Output:
```
[1, 10, 2, 3]
```

##➖ Removing Elements
###remove()

Removes a specific value.

```python
numbers = [1, 2, 3, 4]
numbers.remove(2)
print(numbers)
```

Output:
```
[1, 3, 4]
```

###pop()

Removes element by index.

```python
numbers = [1, 2, 3, 4]
numbers.pop()
print(numbers)
```

Output:
```
[1, 2, 3]
```

````python
numbers = [1, 2, 3]
numbers.append(4)
print(numbers)
````

````python
numbers = [1, 2, 3, 4]
numbers.remove(2)
print(numbers)
````
