# 📦 Tuples

A **tuple** is very similar to a list, but it has one important difference:

> Tuples are **immutable**, meaning they **cannot be changed after creation**.

Tuples use **parentheses `()`**.

### Example

```python
coordinates = (10, 20)

print(coordinates)
print(type(coordinates))
```
Output:
```
(10, 20)
<class 'tuple'>
```
Once created, the values inside a tuple cannot be modified.

````python
coordinates = (10, 20)

print(coordinates)
print(type(coordinates))
````

---
#🔎 Accessing Tuple Elements

Tuple elements can be accessed using indexing, just like lists.

**Remember: Python indexing starts at 0**

Example
```python
colors = ("red", "green", "blue")

print(colors[0])
print(colors[2])
```
Output:
```
red
blue
```
###Negative Indexing

We can also access elements from the end of the tuple.
```python
print(colors[-1])
```
Output:
```
blue
```

````python
colors = ("red", "green", "blue")

print(colors[0])
print(colors[2])
print(colors[-1])
````

---
#🔒 Why Use Tuples?

Tuples are useful when the data should not change.

Examples include:

* 📍 geographic coordinates
* 📅 dates
* ⚙️ configuration settings
* 🧠 fixed machine learning parameters

###Example — Storing Location Coordinates
```python
location = (40.7128, -74.0060)

print("Latitude:", location[0])
print("Longitude:", location[1])
```
Output:
```
Latitude: 40.7128
Longitude: -74.0060
```
Since coordinates should remain constant, tuples are a safe and reliable structure.

````python
location = (40.7128, -74.0060)

print("Latitude:", location[0])
print("Longitude:", location[1])
````

#🔗 Sets

A set is a collection of unique elements.

Key characteristics of sets:
* elements are unordered
* duplicates are automatically removed
* very useful for data cleaning

Sets use curly braces `{}`.

Example
```python
numbers = {1, 2, 3, 3, 4, 5}
print(numbers)
```
Output:
```
{1, 2, 3, 4, 5}
```
Notice that the duplicate 3 appears only once.

````python
numbers = {1, 2, 3, 3, 4, 5}

print(numbers)
````

---
#➕ Adding Elements to a Set

We can add new values using the `add()` method.

Example
```python
fruits = {"apple", "banana"}
fruits.add("mango")
print(fruits)
```

Output (order may vary):
```
{'apple', 'banana', 'mango'}
```
Because sets are unordered, the order of elements may change.

````python
fruits = {"apple", "banana"}
fruits.add("mango")

print(fruits)
````

---
#➖ Removing Elements from a Set

We can remove elements using the `remove()` method.

Example
```python
fruits = {"apple", "banana", "mango"}
fruits.remove("banana")
print(fruits)
```
Output:
```
{'apple', 'mango'}
```

````python
fruits = {"apple", "banana", "mango"}
fruits.remove("banana")
print(fruits)
````

---
#📊 Why Sets Are Useful

Sets are commonly used when we need to:
* remove duplicate values
* track unique items
* compare collections of data

###For example:

Cleaning duplicates in a dataset:
```python
data = [1, 2, 2, 3, 4, 4, 5]
unique_data = set(data)
print(unique_data)
```
Output:
```
{1, 2, 3, 4, 5}
```
This is very useful in data science workflows.

````python
data = [1, 2, 2, 3, 4, 4, 5]
unique_data = set(data)
print(unique_data)
````

Write a program that:
