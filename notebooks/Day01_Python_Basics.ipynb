## 🖨️ Printing in Python
The `print()` function displays output.

```python
print("Hello AI World")
```

## 📦 Variables

Variables store information in Python.

```python
name = "xyz"
age = 100

print(name)
print(age)
```

## ➕ Basic Math Operations

```python
a = 10
b = 5

print("Addition:", a + b)
print("Multiplication:", a * b)
```

## ⌨️ User Input

The `input()` function lets the user enter information.

```python
name = input("Enter your name: ")
print("Hello", name)
```

## 🧩 Mini Python Program

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))

print("Hello", name)
print("Next year you will be", age + 1)
```
