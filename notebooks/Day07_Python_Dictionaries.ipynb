# 🧠 Introduction to Python Dictionaries

Dictionaries are one of the **most useful data structures in Python**.

Unlike lists (which store values using indexes), dictionaries store data using **key–value pairs**.

This makes dictionaries ideal for storing **structured information**.

Example:

```
Name → Alex
Age → 21
Course → AI
```

In Python, this relationship is stored as:
```
key : value
```

Dictionaries are widely used in:

- 📊 structured datasets
- 🌐 API responses (JSON)
- 🤖 machine learning configurations
- ⚙️ real-world applications

Today we will learn:

- 📦 Creating dictionaries  
- 🔎 Accessing values  
- ➕ Adding new data  
- ⚙️ Dictionary methods

---
---

# 📦 Creating a Dictionary

A dictionary is created using **curly braces `{ }`**.

Inside the dictionary, each item is written as:


key : value


### Example

```python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}
print(student)
```
Output:
```
{'name': 'Alex', 'age': 21, 'course': 'AI'}
```
Each piece of information is stored using a unique key.

````python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

print(student)
````

---
---
#🔎 Accessing Dictionary Values

To access a value in a dictionary, we use its key.

Syntax
```
dictionary[key]
```
Example
```python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

print(student["name"])
print(student["age"])
```
Output:
```
Alex
21
```
Here:

* "name" is the key
* "Alex" is the value

````python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

print(student["name"])
print(student["age"])
````

---
---
#➕ Adding New Data

We can easily add new key–value pairs to a dictionary.

Example
```python
student = {
    "name": "Alex",
    "age": 21
}

student["city"] = "Bangalore"
print(student)
```

Output:
```
{'name': 'Alex', 'age': 21, 'city': 'Bangalore'}
```
This makes dictionaries very flexible and dynamic.

````python
student = {
    "name": "Alex",
    "age": 21
}

student["city"] = "Bangalore"

print(student)
````

#⚙️ Useful Dictionary Methods

Python provides several built-in methods to work with dictionaries.

---
#🔑 Getting All Keys

The keys() method returns all the keys in the dictionary.

```python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}
print(student.keys())
```

Output:
```
dict_keys(['name', 'age', 'course'])
```

````python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

print(student.keys())
````

---
#📊 Getting All Values

The values() method returns all values.
```python
print(student.values())
```
Output:
```
dict_values(['Alex', 21, 'AI'])
```

````python
print(student.values())
````

---
#🧩 Getting Key–Value Pairs

The items() method returns both keys and values.

```python
print(student.items())
```
Output:
```
dict_items([('name', 'Alex'), ('age', 21), ('course', 'AI')])
```
This is useful when looping through dictionaries.

````python
print(student.items())
````

---
#🛡 Safe Access with get()

Using get() prevents errors if a key does not exist.
```python
print(student.get("name"))
print(student.get("city"))
```
If the key is missing, Python returns None instead of an error.

````python
print(student.get("name"))
print(student.get("city"))
````
