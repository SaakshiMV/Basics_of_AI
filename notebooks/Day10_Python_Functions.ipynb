# 🧠 Introduction — Python Functions

Functions allow us to **organize code into reusable blocks**.

Instead of writing the same code multiple times, we can **define it once and reuse it whenever needed**.

Functions are extremely important in:

- machine learning pipelines
- data processing systems
- automation scripts
- large software projects

They help break **large problems into smaller manageable pieces**.

Today we will learn:

- ⚙️ Creating functions  
- 📥 Function parameters  
- ➕ Multiple parameters  
- 🔙 Return values  
- ⚙️ Default parameters

# ⚙️ Basic Function

A function is created using the **`def` keyword**.

### Example

```python
def greet():
    print("Hello! Welcome to Python learning.")

#Calling the function:
greet()
```
Output:
```
Hello! Welcome to Python learning.
```
Functions allow us to reuse logic easily.

````python
def greet():
    print("Hello! Welcome to Python learning.")

greet()
````

#📥 Functions with Parameters

Functions can accept **input values**, called **parameters**.

Example
```python
def greet(name):
    print("Hello", name)

#Calling the function:
greet("Alex")
```
Output:
```
Hello Alex
```
Here, **`name`** is the parameter that receives the input value.

````python
def greet(name):
    print("Hello", name)

greet("xyz")
````

#➕ Function with Multiple Parameters

Functions can take **multiple inputs**.

Example
```python
def add_numbers(a, b):
    result = a + b
    print("Sum:", result)

#Calling the function:
add_numbers(5, 3)
```
Output:
```
Sum: 8
```
This allows functions to perform **more complex operations**.

````python
def add_numbers(a, b):
    result = a + b
    print("Sum:", result)

add_numbers(5, 3)
````

#🔙 Returning Values

Instead of printing results, functions can **return values** using the **`return`** keyword.

This is extremely important because returned values can be **stored, reused, or passed to other functions**.

Example
```python
def multiply(a, b):
    return a * b

#Using the returned value:
result = multiply(4, 5)
print("Result:", result)
```
Output:
```
Result: 20
```

````python
def multiply(a, b):
    return a * b

result = multiply(4, 5)
print("Result:", result)
````

#⚙️ Default Parameters

Functions can have **default values**.

If no argument is provided, Python uses the default value.

Example
```python
def greet(name="User"):
    print("Hello", name)

#Calling the function:
greet()
greet("Alex")
```
Output:
```
Hello User
Hello Alex
```
Default parameters make functions more flexible.

````python
def greet(name="User"):
    print("Hello", name)

greet()
greet("Alex")
````

#🧩 Mini Project — Temperature Converter 🌡️

Build a simple temperature converter using functions.

Program should:

1️⃣ take temperature in Celsius

2️⃣ convert it to Fahrenheit

Formula:
```
F = (C × 9/5) + 32
```

````python
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

temp = float(input("Enter temperature in Celsius: "))
result = celsius_to_fahrenheit(temp)
print("Temperature in Fahrenheit:", result)
````
