# 🧠 Introduction

Python supports different **data types** to store different kinds of information.

---
---
# 🔢 Numbers

Numbers are used for mathematical operations.

Common types:
- `int` → integers
- `float` → decimal numbers

---
## 🔢 Integers (int)

Integers are **whole numbers** without decimals.

Examples:
- Age
- Year
- Count of items

```python
age = 30
year = 2026

print("Age:", age)
print("Year:", year)

print("Type:", type(age))
```

---

## 🔣 Floating Numbers (float)

Floats represent **decimal numbers**.

These are commonly used in **scientific calculations and machine learning models**.

```python
height = 5.7
temperature = 36.5

print("Height:", height)
print("Temperature:", temperature)

print(type(height))
```

```python
a = 10
b = 3.5

print("Integer:", a)
print("Float:", b)
```

---

# 🔤 Strings

Strings are used to store text.

Example:
- names
- sentences
- messages

```python
name = "xyz"
message = "Welcome to AI learning"

print(name)
print(message)
```

---

# 🔍 Checking Data Types

Python provides a function called `type()` to check the data type of a variable.

```python
x = 10
y = 3.14
z = "Hello"

print(type(x))
print(type(y))
print(type(z))
```

---

# ✂️ String Operations

Python allows us to manipulate text easily.

```python
text = "Artificial Intelligence"

print("Uppercase:", text.upper())
print("Lowercase:", text.lower())
print("Length:", len(text))
```

---

# 📦 Lists

Lists store multiple values in a single variable.

Example:
- dataset columns
- image pixels
- collections of numbers

```python
numbers = [1, 2, 3, 4, 5]

print(numbers)
```

---

# 🔎 Accessing Elements

Lists use **indexing**.

Index starts at **0**.

```python
numbers = [10, 20, 30, 40]

print(numbers[0])
print(numbers[2])
```

---

# 🔎 Adding elements to a list

```python
numbers = [1, 2, 3]
numbers.append(4)
print(numbers)
```

---

# 📍 Tuples

Tuples are similar to lists but **cannot be changed after they are created**.

They are called **immutable data structures**.

Tuples are useful for storing **fixed values** such as coordinates.

```python
coordinates = (10, 20)

print(coordinates)
print(type(coordinates))
```

---

# 🔗 Sets

Sets store **unique values only**.

Duplicate values are automatically removed.

```python
numbers = {1, 2, 3, 3, 4, 5}

print(numbers)
```

---

# 🗂️ Dictionaries

Dictionaries store **data in key-value pairs**.

This structure is widely used in **data science and APIs**.

```python
student = {
    "name": "xyz",
    "age": 30,
    "course": "AI"
}

print(student)
```

```python
print(student["name"])  #used for accessing data in a dictionary
```

# 🧩 Challenge — Build a Mini AI Learner Profile

Write a program that:

1️⃣ asks the user for their **name**  
2️⃣ asks for their **age**  
3️⃣ asks for their **favorite programming language**  
4️⃣ asks for **3 favorite AI tools**

Store the information using:
- **strings**
- **list**
- **dictionary**

Finally, print the complete **AI learner profile**.

```python
# Get user input
name = input("Enter your name: ")
age = int(input("Enter your age: "))
language = input("Enter your favorite programming language: ")

tool1 = input("Enter AI tool 1: ")
tool2 = input("Enter AI tool 2: ")
tool3 = input("Enter AI tool 3: ")

# Store AI tools in a list
ai_tools = [tool1, tool2, tool3]

# Create dictionary profile
profile = {
    "name": name,
    "age": age,
    "favorite_language": language,
    "ai_tools": ai_tools
}

# Print result
print("\nAI Learner Profile")
print(profile)
```
