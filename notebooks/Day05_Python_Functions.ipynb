# 🧠 Introduction for Functions in Python

As programs grow larger, writing everything in one place becomes **hard to manage**.

Functions help us **organize code into reusable blocks**.

Instead of repeating the same code many times, we can:

1️⃣ Write the code once  
2️⃣ Put it inside a function  
3️⃣ Call the function whenever we need it

Functions make programs:

- cleaner
- easier to maintain
- reusable
- modular

Functions are heavily used in **data science, machine learning, and AI pipelines**.

For example:

- preprocessing data
- training models
- evaluating predictions
- building reusable utilities

Today we will learn:

- ⚙️ Defining functions  
- 📥 Function parameters  
- 🔙 Returning values

# ⚙️ Defining a Function

In Python, functions are created using the **`def` keyword**.

### Syntax

```python
def function_name():
    code_to_run
```
Example
```python
def greet():
    print("Hello, welcome to AI learning!")

greet()
```
How it works

1️⃣ The function greet() is defined

2️⃣ The function runs only when we call it

Output:
```
Hello, welcome to AI learning!
```
This allows us to reuse the same functionality anywhere in the program.

---

````python
def greet():
    print("Hello, welcome to AI learning!")

greet()
````

#📥 Functions with Parameters

Functions can accept inputs, called parameters.

Parameters allow the function to work with different values.

Syntax
```python
def function_name(parameter):
    code
```
Example
```python
def greet(name):
    print("Hello", name)

greet("xyz")
```
Output:
```
Hello xyz
```
Here:

* name is a parameter

* "xyz" is an argument passed to the function

This makes the function dynamic and reusable.

---

````python
def greet(name):
    print("Hello", name)

greet("xyz")
````

#🔙 Returning Values

Functions can send results back using the return keyword.

Returning values allows us to store results and use them later.

Syntax
```python
def function_name():
    return value
```
Example
```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```
Output:
```
8
```
How it works

1️⃣ Function receives 5 and 3

2️⃣ Adds them

3️⃣ Returns the result

4️⃣ Result is stored in the variable result


Returning values is very important for data processing and machine learning workflows.

---

````python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
````

#🧩 Challenge — Simple Calculator

Let's create a small calculator function.

Problem

Write a function that:

1️⃣ Takes two numbers

2️⃣ Returns their sum

Solution
```python
def add_numbers(a, b):
    return a + b

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

result = add_numbers(x, y)

print("Sum:", result)
```
Example interaction:
```
Enter first number: 5
Enter second number: 7
Sum: 12
```
This demonstrates how functions can combine user input and computation.

---

````python
def add_numbers(a, b):
    return a + b

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

result = add_numbers(x, y)

print("Sum:", result)
````
