# 🧠 Introduction for Loops in Python

In programming, we often need to **repeat the same action multiple times**.

Instead of writing the same code again and again, we can use **loops**.

Loops make programs:

- cleaner
- faster to write
- easier to maintain

They are extremely important in **data science and AI**, because they help process **large datasets**.

### Example use cases

Loops are commonly used for:

- 📊 processing datasets
- 🤖 training machine learning models
- 🔁 running simulations
- ⚙️ automating repetitive tasks
- 📂 iterating through files or records

Today we will learn:

- 🔁 `for` loops  
- 🔄 `while` loops

# 🔁 The `for` Loop

A **for loop** is used to **iterate over a sequence of items**.

A sequence can be:

- a list
- a string
- a tuple
- a range of numbers

### Syntax

```python
for variable in sequence:
    code_to_run
```
Example

```python
fruits = ["apple", "banana", "mango"]
for fruit in fruits:
    print(fruit)
```
How it works

Python will:

1️⃣ take the first item → "apple"

2️⃣ print it

Then repeat for each item.


Output:
```
apple
banana
mango
```
This allows us to process collections of data easily.

---

````python
fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)
````

#🔄 The while Loop

A while loop runs as long as a condition is True.

It keeps repeating until the condition becomes False.

###Syntax
```python
while condition:
    code_to_run
```
Example

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

How it works

1️⃣ Start with count = 1

2️⃣ Print the number

3️⃣ Increase count by 1

4️⃣ Repeat until count > 5

Output:
```
1
2
3
4
5
```
⚠️ Important:

If the condition never becomes False, the loop becomes an infinite loop.

---

````python
count = 1

while count <= 5:
    print(count)
    count += 1
````

#🔢 Using range()

The range() function generates a sequence of numbers.

It is commonly used with for loops.

###Basic Syntax
```python
range(stop)
```
Example:
```python
for i in range(5):
    print(i)
```
Output:
```
0
1
2
3
4
```

###Range with Start and Stop
```python
range(start, stop)
```
Example:
```python
for i in range(1, 6):
    print(i)
```
Output:
```
1
2
3
4
5
```

###Range with Step
```python
range(start, stop, step)
```
Example:
```python
for i in range(0, 10, 2):
    print(i)
```
Output:
```
0
2
4
6
8
```

---

````python
for i in range(5):
    print(i)
````
