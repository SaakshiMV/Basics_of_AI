#🔍 The if Statement

The `if` statement checks a **condition**.

If the condition is **True**, the code inside the `if` block runs.  
If the condition is **False**, the code is skipped.

### Syntax

```python
if condition:
    code_to_run
```

⚠️ Important: Python uses **indentation (spaces)** to define blocks of code.

### Example

```python
age = 18

if age >= 18:
    print("You are an adult")
```

### How it works

1. Python checks the condition `age >= 18`
2. Since the condition is **True**, the message is printed

Output:

```
You are an adult
```
---

````python
age = 18

if age >= 18:
    print("You are an adult")
````

# 🔀 `if` and `else`

Sometimes we want the program to run **one block of code if the condition is true**  
and **another block if it is false**.

This is where the **`else` statement** is used.

### Syntax

```python
if condition:
    code_if_true
else:
    code_if_false
```

### Example

```python
age = 15

if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
```

### How it works

- If the condition is **True** → first block runs
- If the condition is **False** → `else` block runs

Output:

```
You are a minor
```
---

````python
age = 15

if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
````

# 🔁 The `elif` Statement

Sometimes we need to check **multiple conditions**.

For example:

- A student might get **A, B, or C**
- A temperature might be **hot, warm, or cold**

Python allows this using **`elif` (else if)**.

### Syntax

```python
if condition1:
    code
elif condition2:
    code
else:
    code
```

### Example

```python
score = 85

if score >= 90:
    print("Grade A")
elif score >= 70:
    print("Grade B")
else:
    print("Grade C")
```

### How it works

Python checks conditions **from top to bottom**:

1️⃣ Is `score >= 90`? → No  
2️⃣ Is `score >= 70`? → Yes  
3️⃣ Print **Grade B**

Output:

```
Grade B
```
---

````python
score = 85

if score >= 90:
    print("Grade A")
elif score >= 70:
    print("Grade B")
else:
    print("Grade C")
````

# ⚖️ Comparison Operators

Comparison operators are used to **compare two values**.

They return **True or False**, which helps Python decide which code to run.

| Operator | Meaning               | Example |
|--------|---------------------|--------|
| `==` | equal to | `a == b` |
| `!=` | not equal | `a != b` |
| `>` | greater than | `a > b` |
| `<` | less than | `a < b` |
| `>=` | greater than or equal | `a >= b` |
| `<=` | less than or equal | `a <= b` |

These operators are very important for **logic and decision-making in programs**.


# 🧪 Comparison Operators Example

```python
a = 10
b = 5

print("a == b:", a == b)   # equal to
print("a != b:", a != b)   # not equal to
print("a > b:", a > b)     # greater than
print("a < b:", a < b)     # less than
print("a >= b:", a >= b)   # greater than or equal
print("a <= b:", a <= b)   # less than or equal
```

Output:

```
a == b: False
a != b: True
a > b: True
a < b: False
a >= b: True
a <= b: False
```
---

````python
# Comparison Operators Example

a = 10
b = 5

print("a == b:", a == b)   # equal to
print("a != b:", a != b)   # not equal to
print("a > b:", a > b)     # greater than
print("a < b:", a < b)     # less than
print("a >= b:", a >= b)   # greater than or equal to
print("a <= b:", a <= b)   # less than or equal to
````

# 🧩 Challenge — Grade Checker

Let's build a small program using **conditional statements**.

### Problem

Write a program that:

1️⃣ Asks the user for their **exam score**

2️⃣ Prints the **grade**

| Score | Grade |
|------|------|

    print("Grade A")
