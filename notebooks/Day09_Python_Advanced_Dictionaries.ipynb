# 🧠 Introduction to Advanced Dictionaries

Dictionaries are one of the **most powerful and commonly used data structures in Python**.

They store data using **key–value pairs**, which makes them perfect for handling **structured information**.

Example:
```
Name → Alex
Age → 21
Course → AI
```

Dictionaries are widely used in:

- 🌐 APIs and JSON data
- 📊 Machine learning datasets
- ⚙️ Configuration settings
- 🧠 Data processing pipelines

Today we will learn:

- ✏️ Updating dictionary values  
- ➕ Adding new key–value pairs  
- ⚙️ Dictionary methods  
- 🔁 Looping through dictionaries  

---
---

# 📦 Creating a Dictionary

A dictionary is created using **curly braces `{}`**.

Each item follows the format:

```
key : value
```

### Example

```python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

print(student)
```
Output:
```
{'name': 'Alex', 'age': 21, 'course': 'AI'}
```

````python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}
print(student)
````

---
#✏️ Updating Dictionary Values

We can update a value by assigning a new value to an existing key.

Example
```python
student = {
    "name": "Alex",
    "age": 21
}

student["age"] = 22
print(student)
```
Output:
```
{'name': 'Alex', 'age': 22}
```
This is useful when modifying or correcting stored information.

````python
student = {
    "name": "Alex",
    "age": 21
}
student["age"] = 22

print(student)
````

---
#➕ Adding New Key–Value Pairs

We can add new data simply by assigning a value to a new key.

Example
```python
student = {
    "name": "Alex",
    "age": 21
}

student["city"] = "Bangalore"
print(student)
```
Output:
```
{'name': 'Alex', 'age': 21, 'city': 'Bangalore'}
```
Dictionaries are dynamic, meaning new information can be added anytime.

````python
student = {
    "name": "Alex",
    "age": 21
}
student["city"] = "Bangalore"

print(student)
````

#⚙️ Useful Dictionary Methods

Python provides several built-in methods for working with dictionaries.

---
#🔑 Getting Keys

Returns all dictionary keys.
```python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}
print(student.keys())
```

Output:
```
dict_keys(['name', 'age', 'course'])
```

````python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

print(student.keys())
````

---
#📊 Getting Values

Returns all values.
```python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}
print(student.values())
```
Output:
```
dict_values(['Alex', 21, 'AI'])
```

````python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

print(student.values())
````

---
#🔗 Getting Key–Value Pairs

Returns both keys and values.
```python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

print(student.items())
```
Output:
```
dict_items([('name', 'Alex'), ('age', 21), ('course', 'AI')])
```

````python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

print(student.items())
````

---
#🛡 Safe Access with get()

The **`get()`** method safely retrieves values without causing an error if the key does not exist.
```python
print(student.get("name"))
print(student.get("city"))
```
If the key does not exist, Python returns None instead of an error.

---
#🔁 Looping Through a Dictionary

We can loop through dictionary items using **`items()`**.

Example
```python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

for key, value in student.items():
    print(key, ":", value)
```
Output:
```
name : Alex
age : 21
course : AI
```
This is very useful when processing structured data.

````python
student = {
    "name": "Alex",
    "age": 21,
    "course": "AI"
}

for key, value in student.items():
    print(key, ":", value)
````
